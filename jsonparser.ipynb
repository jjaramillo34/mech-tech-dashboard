```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

doc = {}
data = []

for i in range(0, 114, 6):
    print(i)
    link = "https://todosporpuertorico.pr.gov/articulos/?start=" + str(i)
    print(link)

    r = requests.get(link)
    print(r.status_code)
    soup = BeautifulSoup(r.content, 'html.parser')
    
    for item in soup.find_all('article', attrs={'class':'blog-item'}):
        doc['title'] = item.find('a').text.strip()
        doc['link'] = item.find('a')['href']
        doc['slug'] = item.find('a')['href'].split('/')[-2]
        doc['figure'] = item.find('figure').find('img')['src']
        doc['blog_texto'] = item.find('div', attrs={'class':'blog-texto'}).text.strip()
        url_blog_detail = f"https://todosporpuertorico.pr.gov/{doc['link']}"
        r1 = requests.get(url_blog_detail)
        soup = BeautifulSoup(r1.content, 'html.parser')
        try:
            doc['entry_title'] = soup.find('h1', attrs={'class':'entry-title'}).text.strip()
        except Exception as e:
            doc['entry_title'] = None
        try:
            doc['entry_image'] = soup.find('div', attrs={'class':'entry-foto'}).find('img')['src']
        except Exception as e:
            doc['entry_image'] = None
        try:
            doc['entry_date'] = soup.find('div', attrs={'class':'h5 text-left'}).text.strip()
        except Exception as e:
            doc['entry_date'] = None
        try:
            doc['encode_content'] = soup.find('div', attrs={'class':'bootstrap-wrapper mt-3'}).find('p').text.encode('utf-8', 'ignore').decode('utf-8', 'ignore')
        except Exception as e:
            doc['encode_content'] = None
        try:
            doc['encode_content_html'] = soup.find('div', attrs={'class':'bootstrap-wrapper mt-3'})
        except Exception as e:
            doc['encode_content_html'] = None
        
        #print(url_blog_detail)
        #print(doc)
        data.append(doc)
        doc = {}
        
df = pd.DataFrame(data)
df.to_csv('todos_por_puerto_rico_blog_v5.csv', index=True, encoding='utf-8')
```
